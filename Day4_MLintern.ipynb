```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

# Load the image
image = cv2.imread('/content/fd.jpeg', cv2.IMREAD_GRAYSCALE)

# Check if the image is loaded
if image is None:
    print("Error: Image not found.")
    exit()

# Initialize ORB detector
orb = cv2.ORB_create()

# Detect keypoints and compute descriptors
keypoints, descriptors = orb.detectAndCompute(image, None)

# Draw keypoints on the image
output_image = cv2.drawKeypoints(image, keypoints, None, color=(0, 255, 0), flags=0)

# Show the image with keypoints using cv2_imshow
cv2_imshow(output_image)

# Wait for a key press (this is useful for local environments, but not needed in Colab)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

# Print the number of keypoints and the shape of descriptors
print(f'Number of Keypoints: {len(keypoints)}')
print(f'Descriptor Shape: {descriptors.shape}')
```

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

# Load the image
image = cv2.imread('/content/fd.jpeg', cv2.IMREAD_COLOR)

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Initialize the ORB detector
orb = cv2.ORB_create()

# Detect keypoints and descriptors
keypoints, descriptors = orb.detectAndCompute(gray, None)

# Draw the keypoints on the image
image_with_keypoints = cv2.drawKeypoints(image, keypoints, None, color=(0, 255, 0), flags=0)

# Display the image with keypoints
cv2_imshow(image_with_keypoints)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
